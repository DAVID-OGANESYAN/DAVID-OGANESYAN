```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

file_path = 'WorldHappiness.xls'
data = pd.read_excel(file_path).dropna()
```

```python
#matrix
numeric_data = data.select_dtypes(include=[np.number])
correlation_matrix = numeric_data.corr()
print(correlation_matrix)
```

```python
#heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Happiness and Other Features')
plt.show()
```

By looking at this correlation matrix, we can see how various factors are related to happiness levels. Each element of the matrix shows the correlation coefficient between the two factors, and the color indicates the strength and direction of this correlation. For example, red shows a strong positive correlation, and blue shows a strong negative one.
# ______________________________________________________

### Strong positive correlation:

1. **Life Ladder and Log GDP per capital**:
- Strong positive correlation (0.79). This means that in countries with high GDP per capita, people are more likely to be happy. Financial stability plays an important role in life satisfaction.

2. **Life Ladder and Social support**:
- Correlation at 0.72. People feel happier when they have the support of friends, family, and the community.

3. **Life Ladder and Healthy life expectation at birth**:
- Positive correlation (0.73). Health is an important factor of happiness  the longer and healthier life, the more life satisfaction.

4. **Log GDP per capital and Healthy life expectancy at birth**:
- Very strong positive correlation (0.83). High GDP is often associated with better medical care and a higher standard of living.

# _________________________________GRAPHS LOWER


### Negative correlations:

1. **Perceptions of corruption**:
- Negative correlation with **"Life Ladder"** (-0.45) and **"Freedom to make life choices"** (-0.47). The higher the perception of corruption, the lower the level of happiness and sense of freedom. People are less happy in countries where they believe that the government and institutions are corrupt.

2. **Negative affect**:
- Negative correlation with **"Life Ladder"** (-0.35) and **"Social support"** (-0.46). Negative emotions reduce the level of happiness and social support. People who experience more negative emotions tend to be less satisfied with their lives.

# ______________________________________________________


### Conclusion:

This matrix clearly shows that factors such as economic well-being, social support, and health strongly influence happiness levels. At the same time, a high level of perception of corruption and frequent negative emotions reduce overall life satisfaction. These relationships highlight how important different aspects of life are for achieving overall well-being.

```python
social_support = data['Social support']
log_gdp = data['Log GDP per capita']
life_ladder = data['Life Ladder']
healthy_life_exp = data['Healthy life expectancy at birth']
```

```python
plt.scatter(log_gdp, life_ladder, alpha=0.6)
plt.title('Life Ladder vs Log GDP per capita')
plt.xlabel('Log GDP per capita')
plt.ylabel('Life Ladder')
plt.grid(True)
plt.show()

print("Higher GDP per capita generally correlates with higher happiness scores.")
```

```python
plt.scatter(social_support, life_ladder, alpha=0.6)
plt.title('Life Ladder vs Social Support')
plt.xlabel('Social Support')
plt.ylabel('Life Ladder')
plt.grid(True)
plt.show()

print("Strong social support correlate with higher happiness scores.")
```

```python
plt.scatter(healthy_life_exp, life_ladder, alpha=0.6)
plt.title('Life Ladder vs Healthy Life Expectancy')
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Life Ladder')
plt.grid(True)
plt.show()

print("Higher healthy life expectancy is associated with higher happiness scores.")
```

```python
plt.scatter(log_gdp, healthy_life_exp, alpha=0.6)
plt.title('Log GDP per capita vs Healthy Life Expectancy')
plt.xlabel('Log GDP per capita')
plt.ylabel('Healthy Life Expectancy')
plt.grid(True)
plt.show()

print("Higher GDP per capita is strongly correlated with longer healthy life expectancy.")
```
